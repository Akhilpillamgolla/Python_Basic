### Print Formating

We can add data to string using + . but that is a lengthy process. So, we have 2 methods for that : 
.format ()  & f-string. Where  format ()  - method | f-string – formatted string literal. 

```python
print('This is string {}'.format('Inserted vales')) # Format method() 
# here {} specifies the position where inserted values must be placed.
```

```python
print('This is string {}{}{}'.format('Inserted vales 1 ','Inserted vales 2 ','Inserted vales 3')) 
```

```python
#If you want to rearrange them, you can by specifying the position in {} 
print('This is string {2}{1}{0}'.format('Inserted vales 1 ','Inserted vales 2 ','Inserted vales 3 ')) 
# remeber the index will be starting from the 0.
```

```python
# we can also use the same index multiple times. 
print('This is string {2}{2}{2}'.format('Inserted vales 1 ','Inserted vales 2 ','Inserted vales 3 ')) 
# remeber the index will be starting from the 0.
```

```python
# we can also bring the value from the else where like 
sum = 10.1 + 40.234 
print('sum of numbers is {}'.format(sum))
```

```python
# Here we can print only the required decimal points using round.
sum = 10.1 + 40.234
print('sum of numbers is {0:1.2f}'.format(sum))
```

```python
# we can also write this in this format
sum = 10.1 + 40.234
print('sum of numbers is {sum:1.1f}'.format(sum=sum)) # as we have same name sum = sum
```
